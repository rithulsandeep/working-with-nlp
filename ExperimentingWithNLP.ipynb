# Default Sentiment Analysis

```python
from transformers import pipeline
classifier = pipeline(task="sentiment-analysis")
result = classifier("Hate this")
print(result)
```

# Sentiment Analysis and Text Classification using specific models

```python
classifier = pipeline(task = "sentiment-analysis", model = "SamLowe/roberta-base-go_emotions", top_k = None)
classifier("Thanks for nothing")
```

```python
classifier = pipeline(task = "text-classification", model = "SamLowe/roberta-base-go_emotions", top_k = None)
classifier("Thanks for nothing")
```

# Text Summarization using Another Model

```python
summarizer = pipeline(task = "summarization", model = "facebook/bart-large-cnn")
text = "The rapid advancement of artificial intelligence has transformed industries worldwide, from healthcare to finance. AI-powered tools now assist doctors in diagnosing diseases with higher accuracy, while algorithms optimize stock market predictions. However, ethical concerns, such as data privacy and job displacement, remain significant challenges. Governments and organizations are working on regulations to ensure responsible AI development. Despite these hurdles, experts agree that AI’s potential benefits outweigh its risks, provided it is implemented thoughtfully."
```

```python
summarized_text = summarizer(text, min_length = 5, max_length = 140)[0]['summary_text']
summarized_text
```
